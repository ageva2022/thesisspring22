```python
import numpy as np 
import matplotlib.pyplot as plt
import random
import csv

from ipynb.fs.full.blb_private import blb
```

```python
def estimator_mean(data):
    return np.mean(data)

def estimator_var(data):
    return np.var(data)
```

```python
header = ['n', 'T', 'true_parameter', 'estimator_mean', 'estimator_variance']

bounds_theta = [180,200]
bounds_sigma_sq = [0,5]

T = 100 #trials
list_of_n = [250, 500, 750, 1000, 2500, 5000]

true_mean = 190
true_sd = 30
true_var = true_sd**2

with open('blb_gaussian_mean_DP1.csv', 'w', encoding='UTF8', newline='') as file:
    writer = csv.writer(file)

    # write the header
    writer.writerow(header)

    for n in list_of_n:
        for t in range(T):
            data = np.random.normal(true_mean, true_sd, n)
            theta, sigma_sq = blb(estimator_mean, data, n, true_var,bounds_theta, bounds_sigma_sq) #returns mean and variance of estimator
            row = [n, t, true_mean, theta, sigma_sq]
        
            # write the data
            writer.writerow(row)
```
