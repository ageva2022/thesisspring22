```python
import numpy as np 
import matplotlib.pyplot as plt
import random
import csv

from ipynb.fs.full.blb_private import blb
```

```python
def estimator_mean(data):
    return np.mean(data)

def estimator_var(data):
    return np.var(data)
```

```python
header = ['n', 'T', 'true_parameter', 'estimator_mean', 'estimator_variance']

bounds_theta = [2,6]
bounds_sigma_sq = [2,6]

T = 100 #trials
list_of_n = [250, 500, 750, 1000, 2500, 5000]

true_theta = 4

with open('blb_poisson_DP.csv', 'w', encoding='UTF8', newline='') as file:
    writer = csv.writer(file)

    # write the header
    writer.writerow(header)

    for n in list_of_n:
        for t in range(T):
            data = np.random.poisson(true_theta, n)
            theta, sigma_sq = blb(estimator_mean, data, n, true_theta, bounds_theta, bounds_sigma_sq) #returns mean and variance of estimator
            row = [n, t, true_theta, theta, sigma_sq]
        
            # write the data
            writer.writerow(row)
```
